## Question 3

Explain the different types of relationships entities can have in a SQL database. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________

There are one-to-one, one-to-many, and many-to-many relationships in SQL. One-to-one refers to direct relationships between a record in one table and a record in another table, of which is the only realtionship those records have. One-to-many reffers to situations where one record may connect to many other records in another table, where the primary key in one record may be the foreign key for multiple records in a corresponding table. 

A many to many realtionship refers to a complex relationship where many records may correspond with many others.
